```python
import requests
```

```python
issuer_req = requests.get('https://www.thaibma.or.th/issuer/listissuer/home')
issued_req = requests.get('https://www.thaibma.or.th/registeredbond/GetBondOutstandingList/CORP')
```

```python
issuer_list = issuer_req.json()
issued_list = issued_req.json()

for bond in issued_list:
    symbol = ''
    for s in bond['Symbol']:
        try:
            int(s)
            break
        except:
            symbol += s
    
    if symbol == 'AEON':
        symbol = 'AEONTS'
    bond['Ticker'] = symbol

    for issuer in issuer_list:
        if issuer['IssuerAbbrName'] == symbol:
            bond['IssuerAbbrName'] = issuer['IssuerAbbrName']
            bond['IssuerNameEN'] = issuer['IssuerNameEN']
            bond['BusinessSector'] = issuer['BusinessSector']
            bond['BusinessSectorName'] = issuer['BusinessSectorName']
#         if symbol == 'AEON' and bond['IssuerAbbrName'] == 'AEONTS':
#             bond['IssuerAbbrName'] = issuer['IssuerAbbrName']
#             bond['IssuerNameEN'] = issuer['IssuerNameEN']
#             bond['BusinessSector'] = issuer['BusinessSector']
#             bond['BusinessSectorName'] = issuer['BusinessSectorName']
            
print(issued_list)
```

```python
sector = 'PROP'
issued_list2 = []
# print(issued_list)
for bb in issued_list:
    try: 
        if bb['BusinessSector'] == sector:
            issued_list2.append(bb)
    except:
        print(bb['IssueID'])
        
```

```python
issued_list2
```

```python
import pandas as pd
import datetime
```

```python
df = pd.DataFrame(issued_list2)
```

```python
df['MaturityDate'] = pd.to_datetime(df['MaturityDate'])
df['IssuedDate'] = pd.to_datetime(df['IssuedDate'])

df['Tenor'] = df['MaturityDate'] - df['IssuedDate']

df
```

```python
columns_to_drop = ['IssueRating', 'IssueRatingTris', 'IssueRatingFitchth','IssueRatingMoody', 'IssueRatingSP', 'IssueratingFitch', 'IssueratingRI', 'IssueID']
df = df.drop(columns=columns_to_drop)

df
```

```python
df['tenor'] = (pd.to_numeric(df['Tenor'])/10**9)/86400/365

df = df.drop(df[df['tenor'] > 3].index)

df = df.sort_values('InstitutionRating',ignore_index=True)

pd.set_option('display.max_rows', None)

df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
sampledf = df.head(66)

sampledf
```

```python
plt.figure(figsize=(20, 10))
p = sns.regplot(x=sampledf['InterestRate'], y=sampledf['tenor'], ci=None,)
             
plt.xlabel('Coupond (%)')
plt.ylabel('Tenor (year)')

for i in range(sampledf.shape[0]):
    plt.text(sampledf['InterestRate'][i],sampledf['tenor'][i],sampledf['Ticker'][i])
    
slope, intercept, r, p, sterr = scipy.stats.linregress(y=p.get_lines()[0].get_xdata(),
                                                       x=p.get_lines()[0].get_ydata())

print('Rate = ' + str(round(intercept,3)) + ' + ' + str(round(slope,3)) + 'x')
```
